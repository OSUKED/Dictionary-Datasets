```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

import FEAutils as hlp
```

```python
df_cfg_gen = pd.read_csv('../datasets/cfd-generation/actual-cfd-generation-and-avoided-ghg-emissions.csv')

df_cfg_gen.columns = df_cfg_gen.columns.str.lower()
df_cfg_gen['settlement_date'] = pd.to_datetime(df_cfg_gen['settlement_date'], utc=True)
df_cfg_gen = df_cfg_gen.set_index(['settlement_date', 'cfd_id'])
assert df_cfg_gen.index.duplicated().sum() == 0, 'There are duplicate entries in the index'

df_cfg_gen.head(3)
```

```python
s_cfd_payments = (
    df_cfg_gen
    .reset_index()
    .groupby('settlement_date')
    ['cfd_payments_gbp']
    .sum()
)

# Plotting
fig, ax = plt.subplots(dpi=150)

s_cfd_payments.divide(1e6).plot(ax=ax)

hlp.hide_spines(ax)
ax.set_xlabel('')
ax.set_ylabel('CfD Payments (Â£m/MWh)')
```

```python
from collections.abc import Iterable
from sklearn import linear_model

class AxTransformer:
    def __init__(self, datetime_vals=False):
        self.datetime_vals = datetime_vals
        self.lr = linear_model.LinearRegression()
        
        return
    
    def process_tick_vals(self, tick_vals):
        if not isinstance(tick_vals, Iterable) or isinstance(tick_vals, str):
            tick_vals = [tick_vals]
            
        if self.datetime_vals == True:
            tick_vals = pd.to_datetime(tick_vals).astype(int).values
            
        tick_vals = np.array(tick_vals)
            
        return tick_vals
    
    def fit(self, ax, axis='x'):
        axis = getattr(ax, f'get_{axis}axis')()
        
        tick_locs = axis.get_ticklocs()
        tick_vals = self.process_tick_vals([label._text for label in axis.get_ticklabels()])
        
        self.lr.fit(tick_vals.reshape(-1, 1), tick_locs)
        
        return
    
    def transform(self, tick_vals):        
        tick_vals = self.process_tick_vals(tick_vals)
        tick_locs = self.lr.predict(np.array(tick_vals).reshape(-1, 1))
        
        return tick_locs
    
def set_date_ticks(ax, start_date, end_date, axis='y', date_format='%Y-%m-%d', **date_range_kwargs):
    dt_rng = pd.date_range(start_date, end_date, **date_range_kwargs)

    ax_transformer = AxTransformer(datetime_vals=True)
    ax_transformer.fit(ax, axis=axis)
    
    getattr(ax, f'set_{axis}ticks')(ax_transformer.transform(dt_rng))
    getattr(ax, f'set_{axis}ticklabels')(dt_rng.strftime(date_format))

    ax.tick_params(axis=axis, which='both', bottom=True, top=False, labelbottom=True)
    
    return ax
```

```python
fig, ax = plt.subplots(dpi=150)

s_cfd_payments['2017':].resample('M').sum().divide(1e6).plot.bar(ax=ax)

hlp.hide_spines(ax)
ax.set_xlabel('')
ax.set_ylabel('Monthly CfD Payments (Â£m/MWh)')
set_date_ticks(ax, '2017-01-01', '2021-12-01', axis='x', freq='6MS', date_format='%Y-%m')
```

```python
rounded_cost = s_cfd_payments["2021-09":].sum()

# Plotting
fig, ax = plt.subplots(dpi=250)

s_cfd_payments['2021-09':].divide(1e6).plot.bar(ax=ax)

hlp.hide_spines(ax)
ax.set_title(f'Reported CfD Payments for September 2021\nTotal Cost: Â£{rounded_cost:,.0f}')
ax.set_xlabel('')
ax.set_ylabel('Daily CfD Payments (Â£m/MWh)')
set_date_ticks(ax, '2021-09-01', '2021-09-22', axis='x', freq='3d', date_format='%Y-%m-%d')
```
